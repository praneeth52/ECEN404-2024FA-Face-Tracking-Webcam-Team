```python
# YOLOv8 (.s) model

from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the YOLOv8 model
model_path = '/content/content/MyDrive/YOLO/YOLOv8nTraining/facialDetection22/weights/best.pt'
model = YOLO(model_path)

# Path to the test image
image_path = '/content/content/MyDrive/YOLO/testImages/test2.png'

# Perform inference
results = model(image_path)

# Display the results
# You can use YOLO's built-in plotting function
results[0].plot()

# If you prefer OpenCV for further customizations
# (Optional) Load the image with bounding boxes
annotated_image = results[0].plot()

# Show the image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# YOLOv5 (.s) model

import torch
import cv2
import matplotlib.pyplot as plt

# Load the YOLOv5 model
model_path = '/content/content/MyDrive/YOLO/yolov5/runs/train/exp2/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# Path to the test image
image_path = '/content/content/MyDrive/YOLO/testImages/test2.png'

# Perform inference
results = model(image_path)

# Display the results (using YOLOv5's built-in function)
results.print()  # Prints results in the console
results.show()   # Opens an image viewer to display the detections

# (Optional) Save the annotated image
results.save(save_dir='inference_output/')  # Save to 'inference_output/' folder

# If you prefer OpenCV for further customizations
# Load the annotated image for visualization with OpenCV
annotated_image = cv2.imread('inference_output/test2.png')

# Show the image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
